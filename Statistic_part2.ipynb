```python
1.  What is hypothesis testing in statistics?
  -> Hypothesis testing in statistics is a method used to evaluate claims about a population by
    analyzing sample data. It involves formulating a null hypothesis (a statement of no effect or no difference)
    and an alternative hypothesis (a statement that contradicts the null hypothesis).
```

```python
2.  What is the null hypothesis, and how does it differ from the alternative hypothesis?
  ->   The null hypothesis is a statement of "no effect" or "no difference," while the alternative
       hypothesis is a statement that contradicts the null hypothesis, suggesting there
       is a real effect or difference.
   #  Null Hypothesis (H0):
    Assumes no relationship or difference: between variables or groups.
    It's the default position, the status quo that researchers challenge.
    Often involves statements of equality (e.g., =, ≤, ≥).
  # Example:
      "There is no difference in test scores between students who study with flashcards and those who don't
```

```python
3.  What is the significance level in hypothesis testing, and why is it important?
   -> In hypothesis testing, the significance level (denoted by α, alpha) is the
      probability of rejecting the null hypothesis when it is actually true.
    *  It represents the researcher's willingness to accept a Type I error, which is falsely
       rejecting a true null hypothesis.
    # importance :-
# Controlling Type I errors:
    The significance level helps control the risk of making a Type I error.
    A lower significance level (e.g., 0.01) reduces the chance of a false positive but
    may increase the risk of failing to detect a true effect (Type II error).
```

```python
4.  What does a P-value represent in hypothesis testing?
  -> P-value represents the probability of observing results as extreme as, or more extreme
    than, the results actually obtained in a study, assuming the null hypothesis is true
```

```python
5.  How do you interpret the P-value in hypothesis testing?
  -> The P-value is the probability of observing a test statistic as extreme as (or more extreme than)
     the one you calculated assuming the null hypothesis is true.
```

```python
6.  What are Type 1 and Type 2 errors in hypothesis testing?.
  -> A Type I error occurs when a true null hypothesis is incorrectly rejected, leading to a false positive.
    A Type II error happens when a false null hypothesis is not rejected, resulting in a false negative.
```

```python
3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
        x (int): The value of the random variable (0 or 1).
        p (float): The probability of success (0 ≤ p ≤ 1).

    Returns:
        float: The probability P(X = x)
    """
    if x not in [0, 1]:
        return 0.0
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    return p if x == 1 else 1 - p

# Example usage
print("P(X=0):", bernoulli_pdf(0, 0.3))
print("P(X=1):", bernoulli_pdf(1, 0.3))
```

```python
9.  Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def compute_and_plot_z_scores(data):
    """
    Computes Z-scores of the given dataset and plots them.

    Parameters:
    - data (array-like): Input numeric dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot the Z-scores
    plt.figure(figsize=(10, 5))
    plt.plot(z_scores, marker='o', linestyle='-', color='teal')
    plt.axhline(0, color='gray', linestyle='--', label='Mean (Z=0)')
    plt.title('Z-score Plot')
    plt.xlabel('Index')
    plt.ylabel('Z-score')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

    return z_scores

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=100)
z_scores = compute_and_plot_z_scores(sample_data)
```

```python
14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate Z-score of a data point.
    """
    return (x - mean) / std_dev

def plot_standard_normal_with_z(z_score):
    """
    Plot the standard normal distribution and shade the area up to the given Z-score.
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)

    plt.plot(x, y, label='Standard Normal PDF', color='blue')
    plt.fill_between(x, 0, y, where=(x <= z_score), color='lightblue', alpha=0.6, label=f'P(Z ≤ {z_score:.2f})')
    plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
    plt.axvline(0, color='black', linestyle='--', alpha=0.6)

    plt.title('Standard Normal Distribution and Z-score')
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example: Data point, mean, and standard deviation
data_point = 75
mean = 70
std_dev = 10

# Calculate Z-score
z = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point} (mean={mean}, std_dev={std_dev}) is: {z:.2f}")

# Plot standard normal and compare
plot_standard_normal_with_z(z)
```

```python
15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample_data: list or array, the sample observations
    - population_mean: float, the population mean under H0
    - population_std: float, the known population standard deviation
    - alpha: float, significance level (default 0.05)
    - alternative: 'two-sided', 'less', or 'greater'

    Returns:
    - z_stat, p_value, and test result decision
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)

    # Z-statistic
    z_stat = (sample_mean - population_mean) / standard_error

    # P-value based on alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Hypothesis test decision
    decision = "Reject null hypothesis" if p_value < alpha else "Fail to reject null hypothesis"

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision at α = {alpha}: {decision}")

    return z_stat, p_value, decision

# Example: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=50)  # Sample from N(102, 15)

# Hypothesis test: H0: mean = 100, Ha: mean ≠ 100
z_test(sample_data=sample, population_mean=100, population_std=15, alpha=0.05, alternative='two-sided')
```

```python
16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a dataset using the t-distribution.

    Parameters:
    - data: array-like sample
    - confidence: desired confidence level (default 0.95 for 95%)

    Returns:
    - (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-score for two-tailed interval

    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")

    return lower, upper

# Example: Simulated sample data
np.random.seed(1)
sample_data = np.random.normal(loc=100, scale=15, size=40)

# Compute 95% confidence interval
confidence_interval(sample_data)
```

```python
17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate and interpret the confidence interval for the mean of a sample.
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-score for confidence level

    margin_of_error = t_crit * std_err
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    # Output results
    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    print(f"Interpretation: We are {int(confidence * 100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")

    return lower_bound, upper_bound

# Step 1: Generate data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Calculate and interpret the 95% confidence interval
calculate_confidence_interval(data, confidence=0.95)
```

```python
1.  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform a one-sample Z-test.

    Parameters:
    - sample_data: array-like, the sample values
    - population_mean: float, the hypothesized population mean under H0
    - population_std: float, known population standard deviation
    - alpha: float, significance level (default=0.05)
    - alternative: str, 'two-sided', 'greater', or 'less'

    Returns:
    - z_stat, p_value, and decision message
    """
    sample_mean = np.mean(sample_data)
    n = len(sample_data)
    se = population_std / np.sqrt(n)
    z_stat = (sample_mean - population_mean) / se

    # Calculate p-value based on alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'greater', or 'less'")

    # Interpret the result
    if p_value < alpha:
        decision = "Reject the null hypothesis (H₀)"
    else:
        decision = "Fail to reject the null hypothesis (H₀)"

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Significance Level (α): {alpha}")
    print(f"Conclusion: {decision}")

    return z_stat, p_value, decision

# === Example Usage ===
np.random.seed(1)
sample = np.random.normal(loc=102, scale=15, size=50)  # sample with mean ≈ 102

# Hypothesis: Test if mean ≠ 100 (two-sided test)
population_mean = 100
population_std = 15
alpha = 0.05

one_sample_z_test(sample, population_mean, population_std, alpha, alternative='two-sided')
```

```python
3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test comparing sample mean to population mean.

    Parameters:
    - sample_data: list or numpy array of sample values
    - population_mean: known population mean under H0
    - population_std: known population standard deviation
    - alpha: significance level (default = 0.05)

    Returns:
    - z_stat: Z-score
    - p_value: two-tailed p-value
    - decision: result of hypothesis test
    """
    sample_mean = np.mean(sample_data)
    n = len(sample_data)
    std_error = population_std / np.sqrt(n)

    # Z-statistic
    z_stat = (sample_mean - population_mean) / std_error

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Decision
    decision = "Reject H₀" if p_value < alpha else "Fail to reject H₀"

    # Output
    print("=== One-Sample Z-Test ===")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Population Mean: {population_mean}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Significance Level (α): {alpha}")
    print(f"Decision: {decision}")

    return z_stat, p_value, decision

# === Example Usage ===

# Generate sample data
np.random.seed(0)
sample = np.random.normal(loc=102, scale=15, size=50)  # Sample from N(102, 15^2)

# Define population parameters
population_mean = 100
population_std = 15  # known

# Perform the Z-test
one_sample_z_test(sample, population_mean, population_std)
```

```python
4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a two-tailed Z-test and visualize the decision region.
    """
    sample_mean = np.mean(sample_data)
    n = len(sample_data)
    se = population_std / np.sqrt(n)
    z_stat = (sample_mean - population_mean) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Critical Z values
    z_critical = norm.ppf(1 - alpha / 2)

    # Decision
    decision = "Reject H₀" if abs(z_stat) > z_critical else "Fail to reject H₀"

    # Print results
    print("=== Two-Tailed Z-Test ===")
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"Critical Z-value (±): ±{z_critical:.4f}")
    print(f"P-value: {p_value:.4f}")
    print(f"Decision: {decision}")

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color='blue')

    # Fill rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (α/2)')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)

    # Z-stat line
    plt.axvline(z_stat, color='black', linestyle='--', label=f'Z = {z_stat:.2f}')

    # Labels and grid
    plt.title("Two-Tailed Z-Test Decision Region")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    return z_stat, p_value, decision

# === Example Usage ===
np.random.seed(0)
sample = np.random.normal(loc=102, scale=15, size=50)  # mean ≠ 100
population_mean = 100
population_std = 15
alpha = 0.05

two_tailed_z_test(sample, population_mean, population_std, alpha)
```

```python
5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualize Type I (α) and Type II (β) errors for a two-tailed Z-test.

    Parameters:
    - mu0: Population mean under H0
    - mu1: Actual mean under H1
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical Z-values for alpha (two-tailed)
    z_critical = norm.ppf(1 - alpha / 2)

    # Critical values in original scale
    lower_crit = mu0 - z_critical * se
    upper_crit = mu0 + z_critical * se

    # x-axis range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # PDFs for H0 and H1
    y_h0 = norm.pdf(x, loc=mu0, scale=se)
    y_h1 = norm.pdf(x, loc=mu1, scale=se)

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, y_h0, label='H₀: μ = {}'.format(mu0), color='blue')
    plt.plot(x, y_h1, label='H₁: μ = {}'.format(mu1), color='green')

    # Type I error region (blue tails under H0)
    plt.fill_between(x, 0, y_h0, where=(x < lower_crit) | (x > upper_crit), color='red', alpha=0.3, label='Type I Error (α)')

    # Type II error region (middle under H1)
    plt.fill_between(x, 0, y_h1, where=(x >= lower_crit) & (x <= upper_crit), color='orange', alpha=0.3, label='Type II Error (β)')

    # Decision boundaries
    plt.axvline(lower_crit, color='black', linestyle='--', label='Critical Values')
    plt.axvline(upper_crit, color='black', linestyle='--')

    # Labels and legend
    plt.title('Visualization of Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    # Calculate and print Type II error (β)
    beta = norm.cdf(upper_crit, loc=mu1, scale=se) - norm.cdf(lower_crit, loc=mu1, scale=se)
    power = 1 - beta
    print(f"Type I Error (α): {alpha:.2f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Statistical Power (1 - β): {power:.4f}")

# === Example Usage ===
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

```python
9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as st

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Args:
        data (array-like): The sample data (list or NumPy array).
        confidence (float, optional): The desired confidence level (e.g., 0.95 for 95%). Defaults to 0.95.

    Returns:
        tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """
    a = 1.0 * np.array(data)
    n = len(a)
    if n == 0:
      return (np.nan, np.nan)
    m, se = np.mean(a), st.sem(a)
    h = se * st.t.ppf((1 + confidence) / 2., n-1)
    return m-h, m+h

def explain_significance():
  """
  Explains the significance of a confidence interval.
  """
  print("A confidence interval provides a range of values within which the true population mean is likely to fall, based on a sample.")
  print("For example, a 95% confidence interval means that if you were to repeat the sampling process many times, 95% of the calculated intervals would contain the true population mean.")
  print("It essentially quantifies the uncertainty associated with estimating the population mean from a sample.")
  print("A wider interval indicates greater uncertainty, while a narrower interval suggests a more precise estimate.")

# Example usage:
data = [10, 12, 15, 13, 11, 14, 16, 12, 13, 15]
lower, upper = confidence_interval(data, confidence=0.95)
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

explain_significance()
```

```python
10.  Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats

def calculate_margin_of_error(sample, confidence_level):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
        sample (list or array): Sample data points.
        confidence_level (float): Confidence level (e.g., 0.95 for 95%).

    Returns:
        float: Margin of Error
    """
    n = len(sample)
    sample_std = np.std(sample, ddof=1)           # Sample standard deviation
    standard_error = sample_std / np.sqrt(n)      # Standard error
    t_score = stats.t.ppf((1 + confidence_level) / 2, df=n-1)  # t-critical value

    margin_of_error = t_score * standard_error
    return margin_of_error

# Example usage
if __name__ == "__main__":
    # Sample data
    sample_data = [55, 60, 52, 58, 63, 59, 61, 57, 62, 56]

    # Confidence level
    confidence = 0.95

    # Calculate margin of error
    moe = calculate_margin_of_error(sample_data, confidence)
    print(f"Margin of Error at {int(confidence * 100)}% confidence level: ±{moe:.2f}")
```

```python
11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, likelihood, false_positive, base_rate):
    """
    Implements Bayes' Theorem for binary classification.

    Parameters:
        prior: P(Disease) - prior probability of having the disease
        likelihood: P(Positive | Disease) - true positive rate
        false_positive: P(Positive | No Disease) - false positive rate
        base_rate: P(Not Disease) - complement of prior

    Returns:
        Posterior: P(Disease | Positive)
    """
    # P(Positive) = P(Positive | Disease) * P(Disease) + P(Positive | No Disease) * P(No Disease)
    evidence = likelihood * prior + false_positive * base_rate

    # Bayes' Theorem
    posterior = (likelihood * prior) / evidence
    return posterior

# Given values
P_disease = 0.01           # Prior
P_positive_given_disease = 0.99  # Likelihood
P_positive_given_no_disease = 0.05  # False positive rate
P_no_disease = 1 - P_disease       # Complement of prior

# Apply Bayes' Theorem
posterior = bayes_theorem(P_disease, P_positive_given_disease, P_positive_given_no_disease, P_no_disease)
print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

```python
12.Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
from scipy.stats import chi2_contingency

# Create contingency table (observed frequencies)
# Rows = Gender (Male, Female)
# Columns = Preference (Like, Dislike)
data = np.array([[30, 10],
                 [20, 40]])

# Perform Chi-square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Display results
print("Chi-square Statistic:", round(chi2_stat, 4))
print("Degrees of Freedom:", dof)
print("P-value:", round(p_value, 4))
print("\nExpected Frequencies Table:\n", np.round(expected, 2))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis (variables are dependent)")
else:
    print("\nConclusion: Fail to reject the null hypothesis (variables are independent)")
```

```python
13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a chi-square test of independence.

    Parameters:
    observed (2D list or numpy array): Contingency table of observed frequencies.

    Returns:
    expected (numpy array): Expected frequencies.
    """
    observed = np.array(observed)

    # Compute row and column totals
    row_totals = observed.sum(axis=1).reshape(-1, 1)
    col_totals = observed.sum(axis=0).reshape(1, -1)
    grand_total = observed.sum()

    # Compute expected frequencies
    expected = (row_totals @ col_totals) / grand_total
    return expected

# Example contingency table (observed data)
# Format: rows = categories of variable A, columns = categories of variable B
observed_data = [
    [20, 30],
    [30, 20]
]

# Calculate expected frequencies
expected_freq = calculate_expected_frequencies(observed_data)

# Display results using Pandas
print("Observed Frequencies:")
print(pd.DataFrame(observed_data))

print("\nExpected Frequencies:")
print(pd.DataFrame(expected_freq.round(2)))
```

```python
14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy import stats

# Observed frequencies
observed_data = np.array([25, 30, 20, 15, 10])

# Expected frequencies (e.g., assuming a uniform distribution)
# The sum of expected frequencies should equal the sum of observed frequencies
total_observations = np.sum(observed_data)
num_categories = len(observed_data)
expected_data = np.array([total_observations / num_categories] * num_categories)

# Perform the Chi-Square Goodness-of-Fit Test
chi_square_statistic, p_value = stats.chisquare(f_obs=observed_data, f_exp=expected_data)

print(f"Observed Data: {observed_data}")
print(f"Expected Data: {expected_data}")
print(f"Chi-Square Statistic: {chi_square_statistic}")
print(f"P-value: {p_value}")

# Interpretation based on a significance level of 0.05
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed distribution is significantly different from the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed distribution does not significantly differ from the expected distribution.")
```
